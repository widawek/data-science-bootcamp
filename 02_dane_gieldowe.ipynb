```python
df["Average"] = (df.Open + df.Close) / 2
df.head()
```

```python
df = df.sort_index()  # sortowanie po wartości indexu (np dacie)
```

```python
df.head()
```

```python
df["Close_shift"] = df["Close"].shift(1)
df.head()
```

```python
df["Daily_Change"] = df.Close / df.Close_shift -1
df.head()
```

```python
del df["Daily Change"]
```
